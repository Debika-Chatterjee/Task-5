```python
import pandas as pd  # Importing pandas for data manipulation
import matplotlib.pyplot as plt  # Importing matplotlib for plotting

# Load the CSV file into a DataFrame
df = pd.read_csv('Sales Data.csv')

# Display the first few rows to understand the structure
df.head()
```

```python
df.head(10)
```

```python
# Group by Product and sum the Sales
p = df.groupby('Product')['Sales'].sum()

# Display the result
print(p)
```

```python
# Plotting the sales data
p.plot(kind='bar', title='Total Sales per Product', figsize=(8,5), color='green')

# Labeling the axes
plt.xlabel('Product')
plt.ylabel('Total Sales')

# Adjust layout
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('total_sales_per_product.png')  # You can change the filename and format

# Display the plot
plt.show()
```

```python
# Plotting the sales data
s = df.groupby('City')['Sales'].sum()

# Display the result
print(s)
plt.show()
```

```python
s.plot(kind='bar', title='Total Sales per City', figsize=(8,5), color='blue')

# Labeling the axes
plt.xlabel('City')
plt.ylabel('Total Sales')

# Adjust layout
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('total_sales_per_City.png')  # You can change the filename and format

# Display the plot
```
